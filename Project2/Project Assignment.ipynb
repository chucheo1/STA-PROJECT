## ------------------------ The correlation between xgboost & ffnn---------------------------
> $ H_{0} $: The proportion of correct predictions for the xgboost model is equal to the proportion of correct predictions for the ffnn model.

> $ H_{1} $: The proportion of correct predictions for the xgboost model is not equal to the proportion of correct predictions for the ffnn model.

```python
# Set the random seed for reproducibility.
np.random.seed(130)

# Define the number of permutations (bootstrap samples) to generate.
num_permutations = 10000

# Create an array to store simulated differences between models' performance.
simulated_differences = np.array([0.0] * num_permutations, dtype=float)

# Extract the data for XGBoost: y_true_xgboost represents the "Correctness" category.
# Extract the data for FFNN: y_true_ffnn represents the "Correctness" category.
for i in range(num_permutations):
    # Generate bootstrapped samples for XGBoost and FFNN separately.
    # Randomly select indices with replacement to create bootstrapped samples for XGBoost and FFNN.
    xgboost_sampled_idx = df_preds['y_true_xgboost'].sample(frac=1, replace=True).index
    ffnn_sampled_idx = df_preds['y_true_ffnn'].sample(frac=1, replace=True).index

    # Create bootstrapped datasets for XGBoost and FFNN based on the sampled indices.
    xgboost_bootstrapped = df_preds.loc[xgboost_sampled_idx].copy()
    ffnn_bootstrapped = df_preds.loc[ffnn_sampled_idx].copy()

    # Calculate the mean of the "Correctness" category for XGBoost and FFNN bootstrapped samples.
    xgboost_mean = xgboost_bootstrapped['y_true_xgboost'].mean()
    ffnn_mean = ffnn_bootstrapped['y_true_ffnn'].mean()

    # Calculate the difference in means for the bootstrapped samples.
    simulated_differences[i] = xgboost_mean - ffnn_mean

# Calculate the observed test statistic, representing the actual difference in means between XGBoost and FFNN.
observed_test_statistic = df_preds['y_true_xgboost'].mean() - df_preds['y_true_ffnn'].mean()

# Define the null hypothesis (H0): Assuming no difference between the models, the difference in means should be 0.
null_hypothesis = 0

# Calculate the number of simulated differences as extreme or more extreme than the observed test statistic.
num_as_or_more_extreme = (np.abs(np.array(simulated_differences) - null_hypothesis) >= np.abs(observed_test_statistic - null_hypothesis)).sum()

# Calculate the p-value by determining the proportion of extreme simulated differences.
p_value = num_as_or_more_extreme / num_permutations

# Output the calculated p-value.
p_value
```

#### Summary:
> Comparison of Prediction Accuracy Between xgboost and ffnn Models: Lack of significant Differences: With a calculated p-value of 0.9617, the analysis indicates that there is insufficient evidence to reject the null hypothesis. This suggests that there is no significant difference in the proportions of correct predictions between the XGBoost and FFNN models. The high p-value, which exceeds the commonly chosen significance level (alpha) of 0.05, underscores the absence of statistical evidence to support the notion that one model outperforms the other in terms of prediction accuracy.

## Evaluating Economic Inequality Differences between Countries with Varied Human Flight and Brain Drain Scores
> $ H_{0}$:
The mean economic inequality (measured by "fsi_e2:_economic_inequality") is the same for countries with a "fsi_e3:_human_flight_and_brain_drain" score greater than or equal to 8.0 and countries with a score less than 8.0. In other words, there is no difference in economic inequality between these two groups of countries.

> $ H_{1}$:
The mean economic inequality is different for countries with a "fsi_e3:_human_flight_and_brain_drain" score greater than or equal to 8.0 and countries with a score less than 8.0. There is a significant difference in economic inequality between these two groups of countries.

```python
# Filter the dataset to create a subset based on "fsi_e3:_human_flight_and_brain_drain" criteria.
subset_data = df_indicators[df_indicators['fsi_e3:_human_flight_and_brain_drain'] >= 8.0]

# Create the remaining dataset not included in the subset.
remaining_data = df_indicators[df_indicators['fsi_e3:_human_flight_and_brain_drain'] < 8.0]

# Define the number of bootstrap samples to generate.
n_samples = 10000

# Initialize an array to store the mean differences.
mean_differences = np.empty(n_samples)

# Perform bootstrapping to analyze differences in "fsi_e2:_economic_inequality."
for i in range(n_samples):
    # Create bootstrap samples for the subset and remaining data separately.
    subset_sample = np.random.choice(subset_data['fsi_e2:_economic_inequality'], len(subset_data), replace=True)
    remaining_sample = np.random.choice(remaining_data['fsi_e2:_economic_inequality'], len(remaining_data), replace=True)
    mean_differences[i] = np.mean(subset_sample) - np.mean(remaining_sample)

# Calculate the 95% confidence interval for the differences in means.
confidence_interval = np.percentile(mean_differences, [2.5, 97.5])

# Calculate the observed mean difference between the subset and remaining data.
observed_mean_difference = np.mean(subset_data['fsi_e2:_economic_inequality']) - np.mean(remaining_data['fsi_e2:_economic_inequality'])

# Calculate the p-value for the observed mean difference.
p_value = np.mean(mean_differences >= observed_mean_difference)

# Print the results and create a histogram plot with a vertical line for the observed mean difference.
print("95% Confidence Interval:", confidence_interval)
print("Observed Mean Difference:", observed_mean_difference)
print("p-value:", p_value)

# Create a figure for a histogram plot.
fig = px.histogram(mean_differences, nbins=30, labels={'value': 'Mean Differences'}, title='Bootstrap Analysis of "fsi_e2:_economic_inequality"')

# Add a vertical line for the observed mean difference.
fig.add_shape(
    type='line',
    x0=observed_mean_difference,
    x1=observed_mean_difference,
    y0=0,
    y1=0.2,  # Adjust the value for the y-range
    line=dict(color='red', width=2, dash='dash')
)

# Add shaded regions for the confidence interval.
fig.add_shape(
    type='rect',
    x0=confidence_interval[0],
    x1=confidence_interval[1],
    y0=0,
    y1=0.2,  # Adjust the value for the y-range
    line=dict(color='gray'),
    fillcolor='gray',
    opacity=0.3
)

# Show the plot.
fig.show()
```

#### Summary:
> In this analysis, a 95% confidence interval was computed to assess the difference in economic inequality between countries with high and low scores on the "Human Flight and Brain Drain" indicator. The confidence interval ranged from -0.462 to 1.768. The observed mean difference in economic inequality between these two groups of countries was 0.635. 
The p-value obtained was 0.495, suggesting that there is no statistically significant difference in economic inequality between countries with high and low scores on the "Human Flight and Brain Drain" indicator. This means that the observed mean difference is not significantly different from what we would expect to see by random chance alone.
In summary, the data does not provide strong evidence to conclude that there is a significant difference in economic inequality between these two groups of countries based on the "Human Flight and Brain Drain" indicator.
